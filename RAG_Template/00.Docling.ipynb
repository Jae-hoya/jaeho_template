### 한번에 여러개 파일을 처리하기 위한 코드

```python
import os
from docling.document_converter import DocumentConverter

# 변환할 폴더 경로
FILE_PATH = "C:/Users/skyop/jaeho_template/RAG_Template/data"

# DocumentConverter 초기화
converter = DocumentConverter()

# 변환된 결과를 저장할 리스트
results = {}

# 지정된 폴더의 모든 파일 반복
for filename in os.listdir(FILE_PATH):
    if filename.endswith(".pdf"):  # PDF 파일만 처리
        source = os.path.join(FILE_PATH, filename)
        try:
            print(f"Converting: {filename}")
            result = converter.convert(source)
            results[filename] = result  # 결과를 딕셔너리에 저장
        except Exception as e:
            print(f"Error converting {filename}: {e}")

# 결과 확인
for fname, res in results.items():
    print(f"\n==== {fname} ====")
    print(res)  # 필요에 따라 요약, 텍스트 추출 등
```

```python
from langchain_docling import DoclingLoader
from langchain_docling.loader import ExportType
from docling.document_converter import DocumentConverter

# 파싱과 LLM이 이해하기 쉬운 rule_base로 변환까지 하기 때문에 시간이 꽤나 걸린다. 
# FILE_PATH = "https://arxiv.org/pdf/2408.09869" # Docling을 설명하는 논문 - 이미지 경로


FILE_PATH = "C:/Users/skyop/jaeho_template/RAG_Template/data"
FILE_NAME = "LangGraph-Based Integrated Architecture for CrewAI Multi-Agent.pdf"

source = f"{FILE_PATH}/{FILE_NAME}"  # document per local path or URL
converter = DocumentConverter()
result = converter.convert(source)

SAVE_DIR = f"C:/Users/skyop/jaeho_template/RAG_Template/docling_outputs"

loader = DoclingLoader(
    file_path=f"{FILE_PATH}/{FILE_NAME}",
    export_type=ExportType.MARKDOWN, # 이게 없다면 markdown형식이 아닌 그냥 text형식으로 반환
    
)

docs = loader.load()
```

```python
from IPython.display import Markdown

display(Markdown(docs[0].page_content))
```

# md파일로 저장

```python
import os
# 모든 문서를 하나의 txt 파일로 저장
SAVE_DIR = "C:/Users/skyop/jaeho_template/RAG_Template/docling_outputs"

save_path = os.path.join(SAVE_DIR, f"{FILE_NAME}.md")
with open(save_path, "w", encoding="utf-8") as f:
    for i, doc in enumerate(docs, 1):
        f.write(f"## Document {i}\n\n")
        f.write(doc.page_content)
        f.write("\n\n")
```

# JSON파일로 저장(메타데이터가 필요할 때)

```python
import json

os.makedirs(SAVE_DIR, exist_ok=True)

save_path = os.path.join(SAVE_DIR, f"{FILE_NAME}.json")

json_docs = [
    {"content": doc.page_content, "metadata": doc.metadata}
    for doc in docs
]

with open(save_path, "w", encoding="utf-8") as f:
    json.dump(json_docs, f, ensure_ascii=False, indent=2)

print(f"저장 완료: {save_path}")
```
